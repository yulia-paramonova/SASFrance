## Importer les bibliothèques et démarrer

On commence par se connecter au serveur CAS.
Cette session sert d'espace de travail pour charger, explorer et gérer les données selon les droits SAS Viya.

```python
import swat

session = swat.CAS(
    "https://server.demo.sas.com/cas-shared-default-http",
    authinfo='/Users/sbxxab/.authinfo',
    ssl_ca_list="/Users/sbxxab/Downloads/server.demo.sas.com.pem",
)
```

## Vérifier les tables disponibles

```python
session.tableinfo(caslib="casuser")
```

## Charger un CSV depuis le web

CAS peut charger des données depuis différentes sources.
Ici, nous importons un fichier CSV depuis GitHub via HTTP.
La variable *tbl* pointe vers la table CAS chargée.

```python
data_source = "https://raw.githubusercontent.com/sassoftware/sas-viya-programming/master/data/cars.csv"
table_name = "cars"

tbl = session.read_csv(data_source, table_name)
```

## Visualiser rapidement les données

```python
tbl.head(5)
```

## Statistiques de base

```python
tbl.describe()
```

## Statistiques de synthèse

```python
tbl.summary()
```

Avec des paramètres, l'action de synthèse peut cibler uniquement certaines variables et certains indicateurs.

```python
tbl.summary(inputs=["MPG_City", "MPG_Highway"], subSet=["mean", "min", "max"])
```

## Ajouter des variables calculées

```python
tbl.computedVarsProgram = "MPG_Avg = mean(MPG_City, MPG_Highway);"
```

```python
tbl.summary(
    inputs=["MPG_City", "MPG_Highway", "MPG_Avg"], subSet=["mean", "min", "max"]
)
```

## Stocker un résultat

```python
highest_msrp = tbl["MSRP"].max()
highest_msrp
```

## Filtrer avec un paramètre

```python
tbl.query(f"MSRP = {highest_msrp}").head()
```

## Classer les valeurs

```python
df_top10 = tbl.nlargest(10, "MSRP")
display(df_top10)
```
